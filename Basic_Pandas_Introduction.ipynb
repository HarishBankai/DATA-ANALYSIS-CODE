```python
#small code for DataFrame
data = {
    'NAME' : ['A','B','C','D','E'],
    'Division' : ['S1', 'S2', 'S3', 'S4', 'S5'] ,
    'Maths' : [87, 74, 28, 45, 60] ,
    'DSA' : [78, 85, 91, 50, 67] ,
    'FDS' : [65, 90, 67, 59, 78] ,
    'HNR1': [99, 99, 50, 75, 90]
}

df = pd.DataFrame(data)
print(df)
```
